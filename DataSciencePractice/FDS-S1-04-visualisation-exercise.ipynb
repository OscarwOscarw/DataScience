## Question 1.2 (10 points)

Now make an improved version of the plot, so someone presented with it could easily interpret it. There are at least five changes you can make to improve it. Remember that when you export  this notebook to a PDF, it will have a text width of 6 inches.

```python
data = pd.DataFrame(columns=['City', 'V1', 'V2', 'V3'])

plt.figure(figsize=(6, 4))

data.loc[0] = ['London', 5.701, 40, 3.62]
data.loc[1] = ['Edinburgh', 1.830, 6, 4.3]
data.loc[2] = ['Glasgow', 3.400, 5, 1.31]

plt.bar(data['City'], data['V1'], color='blue', label='Population Density')
plt.bar(data['City'], data['V2'], bottom=data['V1'], color='orange', label='Universities')
plt.bar(data['City'], data['V3'], bottom=data['V1'] + data['V2'], color='purple', label='Commuting by bike')

plt.title('Urban Population Density, Number of Universities, and Bicycle Commute Percentage')
plt.ylabel('Value')

plt.legend()

plt.show()
```
